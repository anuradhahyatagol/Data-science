# hierarchial clustering


```python
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
crime = pd.read_csv('crime_data.csv')
```

```python
crime
```

```python
#Normalization  function
def norm_fun(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm = norm_fun(crime.iloc[:,1:])
```

```python
#create dendrogram
dnedogram = sch.dendrogram(sch.linkage(df_norm,method = 'single'))
```

```python
#create Clusters
hc = AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='single')
```

```python
import warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
y_hc = hc.fit_predict(df_norm)
clusters = pd.DataFrame(y_hc , columns=['clusters'])
```

```python
crime['h_clusterid'] = clusters
```

```python
crime['h_clusterid']
```

# K-Means clusters


```python
crime=pd.read_csv('crime_data.csv')
```

```python
crime
```

```python
from sklearn.cluster import KMeans
```

```python
#Normalization Function
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
Scaled_crime_df = scalar.fit_transform(crime.iloc[:,1:])
```

```python
#How to find optimum number of cluster
wcss = []
for i in range(1, 50):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(Scaled_crime_df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 50),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
```

```python
#Build clusters
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(Scaled_crime_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
crime['clusterid_new'] = clusters_new.labels_
```

```python
clusters_new.cluster_centers_
```

```python

crime
```

```python
crime.drop(['a'],axis=1,inplace = True)
```

```python
crime.groupby('clusterid_new').agg(['mean']).reset_index()
```

# DBSCAN Clustering


```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as  plt
```

```python
crime = pd.read_csv('crime_data.csv')
```

```python
crime
```

```python
crime.drop(['a'],axis=1,inplace = True)
```

```python
array = crime.values
```

```python
array
```

```python
stscalar = StandardScaler().fit(array)
x = stscalar.transform(array)
```

```python
x
```

```python
dbscan  = DBSCAN(eps=0.8, min_samples=6)
dbscan.fit(x)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([crime,cl],axis=1)
```
